```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data for illustration
data = {
    'platform': ['Instagram', 'TikTok', 'YouTube'],
    'views': [9200, 9214, 6100],
    'likes': [657, 132, 545],
    'comments': [11, 8, 26],
    'followers': [7, 66, -91],
    'interactions': [695, 694, 626]
}

# Creating the dataframe
df = pd.DataFrame(data)

# Handling infinite values and missing data
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# Drop the 'platform' column for correlation and pairplot
df.drop(columns=['platform'], inplace=True)  # Dropping the non-numeric column

# Correlation matrix
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Pairplot for visualizing relationships between variables
sns.pairplot(df)
plt.show()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
